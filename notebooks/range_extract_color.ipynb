```python
import numpy as np
import cv2
import matplotlib.pyplot as plt


def extract_color_using_range(img_bgr, lower_range=np.array([111, 0, 0]), upper_range=np.array([168,255,211])):
    img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(img_hsv, lower_range, upper_range)
    return mask
```

```python
# hsv range based extraction
lower_range = np.array([110, 15, 0])
upper_range = np.array([150,255,210])
# upper_range = np.array([170,255,210])
img_pth = "../data/cropped/IMG_9770.JPG"
img_bgr = cv2.imread(img_pth)
extracted = extract_color_using_range(img_bgr, lower_range, upper_range)
fig, axes = plt.subplots(figsize=(20, 20))
axes.imshow(extracted, cmap="gray")
```
